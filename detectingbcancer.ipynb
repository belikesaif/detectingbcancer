```python
import numpy as np
from keras.models import Sequential
from keras.layers import LSTM, Dense

# Example data shapes (adjust according to your actual data)
num_samples = 1000
time_steps = 455
num_features = 30
num_classes = 2

x_train = np.random.randn(num_samples, time_steps, num_features)
y_train = np.random.randint(num_classes, size=num_samples)

model = Sequential()
model.add(LSTM(64, input_shape=(time_steps, num_features)))
model.add(Dense(num_classes, activation='softmax'))

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

model.summary()

model.fit(x_train, y_train, epochs=1000)
```
